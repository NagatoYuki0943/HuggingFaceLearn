```python
import torch
from transformers import (
    CLIPConfig,
    CLIPModel,
    CLIPTextConfig,
    CLIPTextModel,
    CLIPVisionConfig,
    CLIPVisionModel,
)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

# CLIPConfig

```python
# Initializing a CLIPConfig with openai/clip-vit-base-patch32 style configuration
config = CLIPConfig()
config
```

```python
model: CLIPModel = CLIPModel(config).to(device)
model
```

```python
model.config
```

# CLIPTextConfig

```python
# Initializing a CLIPText and CLIPVision configuration
text_config = CLIPTextConfig()
text_config
```

```python
text_model: CLIPTextModel = CLIPTextModel(text_config).to(device)
text_model
```

# CLIPVisionConfig

```python
vision_config = CLIPVisionConfig()
vision_config
```

```python
vision_model: CLIPVisionModel = CLIPVisionModel(vision_config).to(device)
vision_model
```

# from_text_vision_configs

```python
config = CLIPConfig.from_text_vision_configs(
    text_config = text_config,
    vision_config = vision_config
)
config
```

```python
model: CLIPModel = CLIPModel(config).to(device)
model
```
