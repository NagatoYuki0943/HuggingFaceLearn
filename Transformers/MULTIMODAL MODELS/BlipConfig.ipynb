```python
import torch
from transformers import (
    BlipConfig,
    BlipModel,
    BlipTextConfig,
    BlipTextModel,
    BlipVisionConfig,
    BlipVisionModel,
)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
device
```

# BlipConfig

```python
# Initializing a BlipConfig with Salesforce/blip-vqa-base style configuration
config = BlipConfig()
config
```

```python
model: BlipModel = BlipModel(config).to(device)
model
```

```python
model.config
```

# BlipTextConfig

```python
# Initializing a BlipTextConfig with Salesforce/blip-vqa-base style configuration
text_config = BlipTextConfig()
text_config
```

```python
text_model: BlipTextModel = BlipTextModel(text_config).to(device)
text_model
```

# BlipVisionConfig

```python
vision_config = BlipVisionConfig()
vision_config
```

```python
vision_model: BlipVisionModel = BlipVisionModel(vision_config).to(device)
vision_model
```

# from_text_vision_configs

```python
config = BlipConfig.from_text_vision_configs(
    text_config=text_config, vision_config=vision_config
)
config
```

```python
model: BlipModel = BlipModel(config)
model
```
